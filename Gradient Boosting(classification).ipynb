```python
# Drop unnecessary columns (e.g., ID)
data = data.drop(columns=["ID"])
```

```python
# Define features and target
features = data[[
    "Age", "Gender", "PAQ605", "Body Mass Index", 
    "Blood Glucose after fasting", "Respondent's Oral", "Blood Insulin Levels"
]]
target = data["Diabetic or not"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train the Gradient Boosting classifier
gb_classifier = GradientBoostingClassifier(random_state=42, n_estimators=100, learning_rate=0.1)
gb_classifier.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
y_pred = gb_classifier.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
# Print results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n")
print(report)
```
